# 1. What is Pandas


> ไลบรารีเสริมสำหรับทำงานกับ Python เพื่อจัดการและวิเคราะห์ข้อมูลทั้งแบบมิติเดียวและหลายมิติ โดยโครงสร้างข้อมูล Pandas ประกอบไปด้วย
  1. Series
  2. DataFrame
  3. Panel

## 1.1 Series


> มีโครงสร้งคล้าย array มิติเดียว เก็บข้อมูลต่างชนิดกันได้



## 1.2 DataFrame


> มีโครงสร้าง array แบบสองมิติ คล้ายตาราง Excel สามารถนำ Series มาประกอบเป็น DataFrame ได้ นิยมใช้ใน Data Science

## 1.3 Panel


> โครงสร้างแบบสามมิติ คล้ายกับมีแผ่น DataFrame ซ้อนทับกันหลายแผ่น ซึ่งจะประกอบไปด้วย 3 แกน ดังนี้
  1. items: axis 0 = DataFrame แต่ละชั้น
  2. major_axis: axis 1 = index หรือ row ของแต่ละ DataFrame
  3. minor_axis: axis 2 = column ของแต่ละ DataFrame

### 3.1.3 การสร้างจาก List/Tuple ที่เป็นเมตริกซ์


> โดยปกติ DataFrame จะแสดง index ที่ column แรก แต่สามารถกำหนด index ใหม่ได้ โดย


```

```





````python
data = [['a',21,100],
        ['b',20,98],
        ['c',22,95],
        ['d',19,100],
        ['e',21,93]]
col = ['Name','Age','Score']

df = pd.DataFrame(data, columns = col)
df
````

````python
df.set_index(['Name'], inplace = True)
df
````

### 3.1.4 การสร้างจาก List โดยใช้ Zip


> บางข้อมูลอาจมี List หลายตัว เช่น อายุ เพศ คะแนน ต้องการรวม List ให้เป็น DataFrame โดยการใช้ zip



```
list(zip(listName))
```





````python
l1 = ['A','B','C']
l2 = [20,43,34]

data = list(zip(l1, l2))
col = ['Name','Score']

df = pd.DataFrame(data, columns = col)
df
````

## 3.2 การสร้าง DataFrame จาก Dictionary


> ความแตกต่างระหว่าง List/Tuple และ Dictionary คือมันจะกำหนดชื่อ Column เองเลย



````python
data = [{'Name':'a', 'Age': 21, 'Score': 100},
        {'Name':'b', 'Age': 20, 'Score': 98},
        {'Name':'c', 'Age': 19, 'Score': 95},
        {'Name':'d', 'Age': 22, 'Score': 92},
        {'Name':'e', 'Age': 21, 'Score': 100}]

df = pd.DataFrame(data)
df
````

## 3.3 การสร้าง DataFrame จาก Series


> หากมีข้อมูล Series อยู่แล้ว สามารถนำมาสร้างเป็น DataFrame ได้เลย โดยการใช้ Dictionary กำหนดชื่อ column



````python
name = ['a','b','c','d','e']
age = [16,17,16,15,17]
score = [100,98,97,100,94]

col = ['Name','Age','Score']

name_series = pd.Series(name)
age_series = pd.Series(age)
score_series = pd.Series(score)

frame = {'Name':name_series, 'Age':age_series, 'Score':score_series}

df = pd.DataFrame(frame)
df
````

# 4. การอ่านและบันทึกไฟล์ข้อมูล


> ข้อมูลที่นำมาใช้งานกับ Data Science ส่วนใหญ่จะเป็นรูปแบบไฟล์ เช่น ฐานข้อมูล, Excel, CSV

## 4.1 การบันทึก DataFrame เป็นไฟล์ CSV



> ไฟล์ CSV เป็นไฟล์แบบ Text ที่คั่นด้วย ( , ) โดยจะต้องใช้คำสั่ง


```
FilePath = 'Filename' # คำสั่งชื่อหรือ path สามารถกำหนดได้ เช่น data/testpd01.csv
DataFrame.to_csv(parameters)
```


> Parameters ประกอบด้วย
  * header ( True/False )
    * ระบุว่าต้องการบันทึกชื่อ column หรือไม่
  * index ( True/False )
    * ระบุว่าต้องการบันทึกเลข index ( column แรก ) ซึ่งตามปกติจะบันทึกถ้าเราไม่กำหนด ( True )
  * index_label ( String )
    * บันทึกชื่อคอลัมน์ของ index ในกรณีที่บันทึก index

````python
data = [{'Name':'a', 'Age': 21, 'Score': 100},
        {'Name':'b', 'Age': 20, 'Score': 98},
        {'Name':'c', 'Age': 19, 'Score': 95},
        {'Name':'d', 'Age': 22, 'Score': 92},
        {'Name':'e', 'Age': 21, 'Score': 100}]

df = pd.DataFrame(data)
````

````python
# บันทึกทั้งหมด แต่ไม่มี index_label

FilePath = 'testpd01.csv'
df.to_csv(FilePath)
````

##5.2 การกำหนดตัวเลือกการอ่านไฟล์ Excel


> กรณีที่อ่านข้อมูลบางส่วน สามารถกำหนดค่าต่างๆได้ดังนี้
  * header ( integer )
    * ระบุว่าข้ามข้อมูลส่วนที่เป็น header กี่แถว
  * usecols ( List หรือ String )
    * ระบุว่าต้องการ column ใดบ้าง
  * names ( List )
    * การตั้งชื่อ column ใหม่
  * index_col ( String)
    * ระบุว่า column ไหนเป็น index
  * sheet name (String)
    * ระบุ worksheet ที่ต้องการ
  * skiprows ( List, Integer )
    * ระบุว่าข้ามแถวไหนบ้าง
  * nrows ( integer )
    * ระบุว่าอ่านกี่แถว
