ใน Python การสร้างฟังก์ชันใช้คีย์เวิร์ด def พร้อมระบุชื่อฟังก์ชันและพารามิเตอร์ (ถ้ามี) ตามด้วยบล็อกของโค้ดที่ฟังก์ชันต้องการทำงาน

ฟังก์ชันสามารถรับค่าผ่านพารามิเตอร์ได้ เพื่อนำมาใช้งานภายในฟังก์ชัน:

````python
def greet(name):
    print(f"Hello, {name}!")
````

````python
greet("Kannapat")
````

ในบางกรณี เราอาจต้องการให้ฟังก์ชันส่งค่ากลับไปยังตัวแปรหรือโค้ดที่เรียกใช้งานฟังก์ชัน ใช้คีย์เวิร์ด return:

````python
def add_numbers(x, y):
    return x + y
````

````python
result = add_numbers(3, 5)
result
````

ฟังก์ชันสามารถรับอาร์กิวเมนต์ที่ไม่จำกัดจำนวนได้ โดยใช้ * หน้าชื่อพารามิเตอร์

````python
def print_fruits(*fruits):
    for fruit in fruits:
        print(fruit)
print_fruits("Apple", "Banana", "Cherry")
````

````python
def print_pyramid(height):
    for i in range(1, height + 1):
        # Print spaces before the numbers
        for j in range(height - i):
            print(" ", end="")
        # Print increasing numbers
        for j in range(i):
            print(i, end="")
        # Move to the next line after each row is printed
        print()

# ให้ความสูง (height) มีค่าเท่ากับ 5
height = 5
print_pyramid(height)
````

## 7.7 การสร้างตัวเลขสุ่ม


> การสร้างตัวเลขสุ่มใช้งาน ด้วย


```
np.random
```

### 7.7.1 การสุ่มตามจำนวนที่ต้องการ


```
np.random.rand(จำนวนเลขสุ่ม) #สร้างเลขสุ่ม 0 - 1
```



````python
np.random.rand(4)
````

### 7.7.2 การกำหนดตัวเลขสุ่ม


```
np.random.rand(จำนวนเลขสุ่ม) * 5 #สร้างตัวเลขสุ่ม 0 - 5
```

หากต้องการกำหนดช่วงเลขสุ่ม


```
np.random.rand(จำนวนเลขสุ่ม) * 30 + 20 #ต้องการเลข 20 - 50 ซึ่ง 20 เป็นค่าต่ำที่สุด และค่าที่สูงที่สุดคือ (30 + 20) 50
```

````python
np.random.rand(4)*10
````

````python
np.random.rand(3) * 5 + 5 #สร้างตัวเลขสุ่ม 5 - 10
````

### 7.7.3 การสร้างตัวเลขสุ่มแบบ Matrix

````python
np.random.rand(5,3)
````

### 7.7.4 การสร้างตัวเลขสุ่มที่มีค่ากระจาย

```
np.random.normal(mean, SD, n)
```

> แล้วกำหนดค่าเฉลี่ย, SD, จำนวนเลขสุ่มที่ต้องการ

````python
mu = 5 #mean
sigma = 0.5 #stadard deviation

s = np.random.normal(mu, sigma, 1000)
s[:5]
````

````python
#ทดสอบแสดงเป็นกราฟ
plt.hist(s, bins=40)
plt.show()
````
