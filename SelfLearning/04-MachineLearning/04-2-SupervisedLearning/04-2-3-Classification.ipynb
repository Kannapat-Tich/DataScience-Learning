## 1.2 Classify by Sepal

```python
# Data
feature = ['sepal length (cm)','sepal width (cm)']

X = iris_df[feature].values
y = iris_df['species'].values
y_names = iris.target_names

# Decision Tree
    # Create Model
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier()
tree.fit(X,y)
    # Juning Parameter
DecisionTreeClassifier(class_weight = 'balanced', criterion = 'gini', max_depth = None,
                       max_features = None, max_leaf_nodes = None,
                       min_impurity_decrease = 0.0, min_samples_split = 2,
                       min_weight_fraction_leaf = 0.0, random_state = None,
                       splitter = 'best') 

# Plot tree
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
plt.figure(figsize = (25, 20))
plot_tree(tree, filled = True, feature_names = feature, 
          class_names = y_names, rounded = True)
plt.show()

# Model Evaluation
    # Create Test set
from sklearn.model_selection import train_test_split
    # Split out 20 %
X_test, X_train, y_test, y_train = train_test_split(X, y, test_size = 0.2)

    # Create & Train
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier()
tree.fit(X_train, y_train)
    # Prediction
y_pred = tree.predict(X_test)
    # Result
print('Model Prediction Result =', y_pred)
print('-----------------------------------------------')
print('Actual Prediction Result =', y_test)

# Evaluate 
from sklearn.metrics import accuracy_score
acc_score1 = accuracy_score(y_test, y_pred)*100
print('-----------------------------------------------')
print('Accuracy Score =', acc_score1,'%')
```

# 2. Naive Bayes Classifier

ใช้หลักการความน่าจะเป็นของเบย์ สมการเดิมคือ
$
P(A | B) = \frac{P(B | A) P(A)}{P(A)}
$

และนำมาปรับรูปแบบเพื่อนำมาใช้งานใน Machine Learning ดังนี้

$$
P(y | X) = \frac{P(X | y) P(y)}{P(X)}
$$

# 3. Support Vector Machine (SVM)
หลักการของ SVM จะนำ Feature space (พื้นที่พล็อตของค่าของกลุ่มข้อมูล) มาหา Hyperplane (แนวที่ใช้ในการแบ่ง class แยกออกจากกัน)

kernel เป็นกลไกเส้นแบ่งกลุ่มข้อมูล ประกอบไปด้วย
1. Linear (kerneal = 'linear')
    * สร้างเส้นแบ่งเป็นเส้นตรง
2. Polynomial 
    * สร้างเส้นแบ่งแบบโค้ง
3. Radial Basis Function Kernel (rbf)
    * สร้างแส้นโค้งและปรับเข้ากับกลุ่มข้อมูลได้ ซึ่งจะนิยมใช้มากที่สุด

```python
# Data
feature = ['sepal length (cm)','sepal width (cm)',
           'petal length (cm)','petal width (cm)']

X = iris_df[feature].values
y = iris_df['species'].values
y_names = iris.target_names

print("Liner :")
# Linear
       # Test set
from sklearn.model_selection import train_test_split
X_test, X_train, y_test, y_train = train_test_split(X, y, test_size = 0.3)
       # Create Model
from sklearn import svm
from sklearn.svm import SVC
linear = svm.SVC(kernel = 'linear', class_weight='balanced')
       # Train
linear.fit(X_train, y_train)
       
from sklearn.model_selection import cross_val_score
cvs = cross_val_score(linear, X, y, cv = 10)

print("Cross Valid = {}" .format(cvs.round(3)))
print("Mean = {:.3f} %" .format(cvs.mean() * 100))

print('----------------------------------')

# Polynomial
print("Polynomial :")
       # Test set
from sklearn.model_selection import train_test_split
X_test, X_train, y_test, y_train = train_test_split(X, y, test_size = 0.3)
       # Create Model
from sklearn import svm
from sklearn.svm import SVC
poly = svm.SVC(kernel = 'poly',degree = 3, class_weight='balanced')
       # Train
poly.fit(X_train, y_train)

from sklearn.model_selection import cross_val_score
cvs = cross_val_score(poly, X, y, cv = 10)

print("Cross Valid = {}" .format(cvs.round(3)))
print("Mean = {:.3f} %" .format(cvs.mean() * 100))


print('----------------------------------')
print("RBF :")

# RBF
       # Test set
from sklearn.model_selection import train_test_split
X_test, X_train, y_test, y_train = train_test_split(X, y, test_size = 0.3)
       # Create Model
from sklearn import svm
from sklearn.svm import SVC
rbf = svm.SVC(gamma = "auto", class_weight='balanced')
       # Train
rbf.fit(X_train, y_train)

from sklearn.model_selection import cross_val_score
cvs = cross_val_score(rbf, X, y, cv = 10)

print("Cross Valid = {}" .format(cvs.round(3)))
print("Mean = {:.3f} %" .format(cvs.mean() * 100))
```

# 4. k-Nearest Neighbors (KNN)
หลักการการทำงานคือการเทียบหาข้อมูลจุดใหม่ ว่ากลุ่มไหนอยู่กับกลุ่มไหน

```python
# Data
feature = ['sepal length (cm)','sepal width (cm)',
           'petal length (cm)','petal width (cm)']

X = iris_df[feature].values
y = iris_df['species'].values

# kNN = 1
print("kNN = 1 :")
    # Create
from sklearn.neighbors import KNeighborsClassifier
    # Train
knn1 = KNeighborsClassifier(n_neighbors = 1)

# CVS 
from sklearn.model_selection import cross_val_score
cvs1 = cross_val_score(knn1, X, y, cv = 10)

print("Cross Validation = ",cvs1.round(3))
print("Mean = {:.3f} %" .format(cvs1.mean() * 100))

print('------------------------------------------------------------------')
# kNN = 3
print("kNN = 3 :")
    # Create
from sklearn.neighbors import KNeighborsClassifier
    # Train
knn3 = KNeighborsClassifier(n_neighbors = 3)

# CVS 
from sklearn.model_selection import cross_val_score
cvs3 = cross_val_score(knn3, X, y, cv = 10)

print("Cross Validation = ",cvs3.round(3))
print("Mean = {:.3f} %" .format(cvs3.mean() * 100))

print('------------------------------------------------------------------')

# kNN = 5
print("kNN = 5 :")
    # Create
from sklearn.neighbors import KNeighborsClassifier
    # Train
knn5 = KNeighborsClassifier(n_neighbors = 5)
# CVS 
from sklearn.model_selection import cross_val_score
cvs5 = cross_val_score(knn5, X, y, cv = 10)

print("Cross Validation = ",cvs5.round(3))
print("Mean = {:.3f} %" .format(cvs5.mean() * 100))
```

# 5. Random Forest
ลักษณะคล้ายกับ Decision Tree แต่จะใช้ลักษณะ Bagging แต่สุ่มเอาข้อมูลตัวเลขเพื่อสร้างต้นไม้หลายๆต้น แล้วจะโหวตผลที่ได้มากที่สุดเป็นผลลัพธ์

```python
# Data
feature = ['sepal length (cm)','sepal width (cm)',
           'petal length (cm)','petal width (cm)']

X = iris_df[feature].values
y = iris_df['species'].values
y_names = iris.target_names

# Test set
from sklearn.model_selection import train_test_split
X_test, X_train, y_test, y_train = train_test_split(X, y, test_size = 0.3)

# Create & Train Model
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier(n_estimators = 80)
forest.fit(X_train, y_train)
    # Juning Parameter
RandomForestClassifier(class_weight = 'balanced',criterion = 'gini',
                          max_depth = None, min_samples_split = 2, 
                          min_samples_leaf = 1, 
                          max_features = 'sqrt')

# Predict
y_pred = forest.predict(X_test)

# Plot one of the trees in the forest
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

plt.figure(figsize = (9, 10))
plot_tree(forest.estimators_[0], 
          filled = True, 
          feature_names = feature, 
          class_names = y_names, 
          rounded = True)
plt.show()

# Evaluate Tester
print('Evaluation Tester')
print('--------------------------------------------')
print("Score = {:.4f} " .format(forest.score(X_test, y_test)))
print('--------------------------------------------')
    # Report
from sklearn.metrics import classification_report
cr = classification_report(y_test, y_pred)
print("classification_report :\n",cr)
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print('--------------------------------------------')
print("confusion_matrix :\n",cm)

# Check False Prediction
import pandas as pd
df = pd.DataFrame()
df['y_test'] = y_test
df['y_pred'] = y_pred
df['Results'] = y_test == y_pred
df[df.Results == False]
```
